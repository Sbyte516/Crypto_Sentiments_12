# News Headlines Sentiment

Use the news api to pull the latest news articles for bitcoin and ethereum and create a DataFrame of sentiment scores for each coin. 

Use descriptive statistics to answer the following questions:
1. Which coin had the highest mean positive score?
2. Which coin had the highest negative score?
3. Which coin had the highest positive score?

```python
# Initial imports
import os
from pathlib import Path
import pandas as pd
from newsapi import NewsApiClient
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from dotenv import load_dotenv
load_dotenv()
analyzer = SentimentIntensityAnalyzer()
%matplotlib inline
```

```python
# Read your api key environment variable
my_key = os.getenv("news_api")
```

```python
my_key
```

```python
# Create a newsapi client
newsapi = NewsApiClient(api_key=my_key)
```

```python
BTC_headlines = newsapi.get_everything(
    q="BTC",
    language="en",
    page_size=100,
    #sort_by="relevancy"
)

# Print total articles
print(f"Total articles BTC: {BTC_headlines['totalResults']}")

# Show sample article
BTC_headlines["articles"][0]
```

```python
# Create the Bitcoin sentiment scores DataFrame

BTC_sentiments = []

for article in BTC_headlines["articles"]:
    try:
        text = article["content"]
        date = article["publishedAt"][:10]
        sentiment = analyzer.polarity_scores(text)
        compound = sentiment["compound"]
        pos = sentiment["pos"]
        neu = sentiment["neu"]
        neg = sentiment["neg"]
        
        BTC_sentiments.append({
            "text": text,
            "date": date,
            "compound": compound,
            "positive": pos,
            "negative": neg,
            "neutral": neu
            
        })
        
    except AttributeError:
        pass
    
# Create DataFrame
BTC_df = pd.DataFrame(BTC_sentiments)

# Reorder DataFrame columns
cols = ["date", "text", "compound", "positive", "negative", "neutral"]
BTC_df = BTC_df[cols]

BTC_df
```

```python
# Fetch the Ethereum news articles
ETH_headlines = newsapi.get_everything(
    q="ETH",
    language="en",
    page_size=100,
    #sort_by="relevancy"
)

# Print total articles
print(f"Total articles ETH: {ETH_headlines['totalResults']}")

# Show sample article
ETH_headlines["articles"][0]
```

```python
# Create the ethereum sentiment scores DataFrame
# Create the Bitcoin sentiment scores DataFrame

ETH_sentiments = []

for article in ETH_headlines["articles"]:
    try:
        text = article["content"]
        date = article["publishedAt"][:10]
        sentiment = analyzer.polarity_scores(text)
        compound = sentiment["compound"]
        pos = sentiment["pos"]
        neu = sentiment["neu"]
        neg = sentiment["neg"]
        
        ETH_sentiments.append({
            "text": text,
            "date": date,
            "compound": compound,
            "positive": pos,
            "negative": neg,
            "neutral": neu
            
        })
        
    except AttributeError:
        pass
    
# Create DataFrame
ETH_df = pd.DataFrame(ETH_sentiments)

# Reorder DataFrame columns
cols = ["date", "text", "compound", "positive", "negative", "neutral"]
ETH_df = ETH_df[cols]

ETH_df
```

# Tokenizer

In this section, you will use NLTK and Python to tokenize the text for each coin. Be sure to:
1. Lowercase each word
2. Remove Punctuation
3. Remove Stopwords

```python
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer, PorterStemmer
from string import punctuation
import string
import re
nltk.download("stopwords")
nltk.download('wordnet')
nltk.download('vader_lexicon')
lemmatizer = WordNetLemmatizer()
```

```python
# Expand the default stopwords list if necessary
```

```python
# First, print default list if stopwords:
print(stopwords.words('english'))
```

```python
# Now, add-on to that default list:
added_words = {'is', 'one', 'was'}
```

```python
def process_text(text):
    stopwords_list = set(stopwords.words('english')) #create stopwords list
    regex = re.compile("[^a-zA-Z]") # Clean up remove things that are not characters
    clean_up = regex.sub('', text) # turn the '' into a space by adding a space
    words = word_tokenize(clean_up)
    lem = [lemmatizer.lemmatize(x) for x in words]
    output = [x.lower() for x in lem if x.lower() not in stopwords_list.union(added_words)]
    text = output.replace("''", " ")
    return text
```

```python

# Create a new tokens column for bitcoin
BTC_df['tokens'] = BTC_df['text'].apply(tokenizer)
BTC_df.head()
```

# NGrams and Frequency Analysis

In this section you will look at the ngrams and word frequency for each coin. 

1. Use NLTK to produce the n-grams for N = 2. 
2. List the top 10 words for each coin. 

```python
from collections import Counter
from nltk import ngrams
```

```python
# Generate the Bitcoin N-grams where N=2
BTC_string = ''.join(BTC_df.text)
BTC_tokenize = tokenizer(BTC_string)
BTCN = Counter(ngrams(BTC_tokenize, n=2))
print(dict(BTCN.most_common(10)))
```

```python
# Generate the Ethereum N-grams where N=2
ETH_string = ''.join(ETH_df.text)
ETH_tokenize = tokenizer(ETH_string)
ETHN = Counter(ngrams(ETH_tokenize, n=2))
print(dict(ETHN.most_common(10)))
```

```python
# Use the token_count function to generate the top 10 words from each coin
def token_count(tokens, N=10):
    """Returns the top N tokens from the frequency count"""
    return Counter(tokens).most_common(N)
```

```python
# Get the top 10 words for Bitcoin
# Get the top 10 words for Bitcoin
BTC_top10 = token_count(BTC_tokenize)
BTC_top10
```

```python
# Get the top 10 words for Ethereum
# Get the top 10 words for Ethereum
ETH_top10 = token_count(ETH_tokenize)
ETH_top10
```

# Named Entity Recognition

In this section, you will build a named entity recognition model for both coins and visualize the tags using SpaCy.

```python
import spacy
from spacy import displacy
```

```python
# Optional - download a language model for SpaCy
# !python -m spacy download en_core_web_sm
```

```python
# Load the spaCy model
nlp = spacy.load('en_core_web_sm')
```

## Bitcoin NER

```python
# Concatenate all of the bitcoin text together
concat_BTC = BTC_df["text"].str.cat()
concat_BTC
```

```python
# Run the NER processor on all of the text
btc_doc = nlp(concat_BTC)
btc_doc
# Add a title to the document
btc_doc.user_data["title"] = "BTC NER"
```

```python
# Render the visualization
colors = {"RED": "linear-gradient(90deg, #aa9cfc, #fc9ce7)"}
options = {"ents": ["RED"], "colors": colors}
displacy.render(btc_doc,style='ent')
```

```python
# List all Entities
for ent in btc_doc.ents:
    print(ent.text, ent.label_)
```
